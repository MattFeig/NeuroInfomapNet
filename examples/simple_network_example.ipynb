```python
import numpy as np
import os, sys 

sys.path.append("../neuroinfomapnet")
from rsfc_tools import create_corrmat
from network_utils import sparsity_inds, apply_sparsity
from community_detection import detect_communities_infomap, clu_to_parcel
```

```python
# LOAD in rsfMRI time-series. Here, we are using the Gordon parcellation. 
# Different parcellations should work with minimal changes

data_dir = 'data'
path_to_read = os.path.join(data_dir, f'example_Gordon333_timeseries.txt')
example_timeseries = np.genfromtxt(path_to_read)
```

```python
# Generate correlation matrix
corr_mat = create_corrmat(example_timeseries)
```

```python
out_dir = 'results'
corr_mat_copy = np.copy(corr_mat)

threshold = .05 # Apply sparsity threshold if desired.
strongest_connections = sparsity_inds(corr_mat_copy,threshold)
sub_thresholded = apply_sparsity(corr_mat_copy, strongest_connections)

# Run infomap
outname = f'pajek_example_thr{threshold}'
detect_communities_infomap(sub_thresholded, outname, out_dir)
```

```python
# Convert clu infomap communty files which infomap outputs into niftis 
# on a cortical surface. Example output saved into examples/results/pajek_example_thr0.05.pscalar.nii

clu_to_parcel(outname,out_dir)
```
